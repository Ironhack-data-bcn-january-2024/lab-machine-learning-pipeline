```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```python
df_train
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
df_train[["fair", "good", "very_good", "premium", "ideal"]] = ohe.fit_transform(df_train["cut"].values.reshape(-1, 1)).toarray()
df_train.drop(columns = ["cut"], inplace=True)
```

```python
df_train[["d", "e", "f", "g", "h", "i", "j"]] = ohe.fit_transform(df_train["color"].values.reshape(-1, 1)).toarray()
df_train.drop(columns = ["color"], inplace=True)
```

```python
df_train[["if", "vvs1", "vvs2", "vs1", "vs2", "si1", "si2", "i1"]] = ohe.fit_transform(df_train["clarity"].values.reshape(-1, 1)).toarray()
df_train.drop(columns = ["clarity"], inplace=True)
```

```python
X_train = df_train.drop(columns=['price'])
y_train = df_train['price']

model = GradientBoostingRegressor()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(df_test)

df_predictions = pd.DataFrame({'id': df_test.index, 'price': predictions})

df_predictions
```

```python
df_predictions.to_csv("guillermo_submission_2.csv", index=False)
```
